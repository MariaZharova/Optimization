# Sequence convergence

1). Show with the definition that the sequence $\left\{ \dfrac{1}{k} \right\}_{k=1}^\infty$ does not have a linear convergence rate (but it converges to zero).

##### Решение:
$$\lim_{k\to\infty}\frac{1}{k} = 0$$

Но несмотря на это, последовательность не сходится линейно; покажем это при помощи теста отношений:

$$\lim_{k\to\infty}\frac{r_{k + 1}}{r_k} = \lim_{k\to\infty}\frac{k}{k + 1} = 1 \Rightarrow\text{сублинейная сходимость.}$$

2). Determine the convergence or divergence of a given sequence $r_{k} = \frac{1}{k^2}$.

##### Решение:

Воспользуемся тестом отношений:

$$\lim_{k\to\infty}\frac{r_{k + 1}}{r_k} = \lim_{k\to\infty}\frac{k^2}{(k + 1)^2} = \lim_{k\to\infty}\frac{k^2}{1 + 2k + k^2} = \lim_{k\to\infty}\frac{1}{\frac{1}{k^2} + \frac{2}{k} + 1} = 1\Rightarrow\text{сублинейная сходимость.}$$

3). Show with the definition that the sequence $\left\{ \dfrac{1}{k^k} \right\}_{k=1}^\infty$ does not have a quadratic convergence rate (but it converges to zero).

$$\lim_{k\to\infty}\frac{1}{k^k} = 0$$

Но по опыту первой задачи мы знаем, что это только необходимое условие сходимости, воспользуемся тестом корней:

$$\lim_{k\to\infty}r_k^{\frac{1}{k}} = \lim_{k\to\infty}\frac{1}{(k^k)^{\frac{1}{k}}} = \lim_{k\to\infty}\frac{1}{k} = 0\Rightarrow\text{сверхлинейная сходимость.}$$

4). Determine the convergence or divergence of a given sequence $r_{k} = 0.707^k$.

##### Решение:

Воспользуемся тестом корней:

$$\lim_{k\to\infty}r_k^{\frac{1}{k}} = \lim_{k\to\infty}(0.707^k)^{\frac{1}{k}} = 0.707 < 1\Rightarrow\text{линейная сходимость.}$$

5). Determine the convergence or divergence of a given sequence $r_{k} = 0.707^{2^k}$.

##### Решение:

Вычислим предел последовательности:

$$\lim_{k\to\infty}0.707^{2^k} = 0.707^{\lim_{k\to\infty}2^k} = 0\text{ - т.к. имеем число от 0 до 1 в бесконечно большой степени.}$$

Для установления типа сходимости, сравним, во сколько раз отличается расстояние до точки сходимости для двух соседних шагов итерационного процесса. Сразу предположим, что сходимость квадратичная и проверим это (т.к. последовательность одномерная, нормой будет обычный модуль):

$$\frac{|0.707^{2^{k + 1}} - 0|}{|0.707^{2^{k}} - 0|^2} = \frac{|0.707^{2^{k + 1}}|}{|0.707^{2^k\cdot2}|} = \frac{|0.707^{2^{k + 1}}|}{|0.707^{2^{k + 1}}|} = 1\Rightarrow\text{действительно, сходимость квадратичная.}$$ 

6). Show that the sequence $x_k = 1 + (0.5)^{2^k}$ is quadratically converged to $1$.

##### Решение:

Вычислим предел последовательности:

$$\lim_{k\to\infty}\left(1 + \left(\frac{1}{2}\right)^{2k}\right) = 1 + 0 = 1$$

Аналогично предыдущему номеру проверим квадратичную сходимость:

$$\frac{|1 + \left(\frac{1}{2}\right)^{2^{k + 1}} - 1|}{|1 + \left(\frac{1}{2}\right)^{2^k} - 1|^2} = \left(\frac{1}{2}\right)^{2^{k + 1} - 2^k\cdot2} = \left(\frac{1}{2}\right)^0 = 1\Rightarrow\text{ квадратичная сходимость.}$$ 

7). Determine the convergence or divergence of a given sequence $r_k =\begin{cases} \left(\frac{1}{4}\right)^{2^k}, & \mbox{if } k\mbox{ is even} \\ \frac{x_{k-1}}{k}, & \mbox{if } k\mbox{ is odd} \end{cases}$.

##### Решение:

Аналогично определим для начала предел последовательности:

$$\lim_{k\to\infty}\frac{r_{k + 1}}{r_k} = \lim_{k\to\infty}\frac{r_k}{r_k\cdot(k + 1)} = \lim_{k\to\infty}\frac{1}{k + 1} = 0\Rightarrow\text{сходимость сверхлинейная.}$$

Теперь для нечётных, тоже применим тест отношений:

$$\lim_{k\to\infty}\frac{r_{k + 1}}{r_k} = \lim_{k\to\infty}\frac{r_{k + 1}\cdot k}{r_{k - 1}} = \lim_{k\to\infty}\frac{\left(\frac{1}{4}\right)^{2^{k + 1}}\cdot k}{\left(\frac{1}{4}\right)^{2^{k - 1}}} = \lim_{k\to\infty}k\cdot\left(\frac{1}{4}\right)^{2^{k + 1} - 2^{k - 1}} = \lim_{k\to\infty}k\cdot\left(\frac{1}{4}\right)^{3\cdot2^{k - 1}} = (*)$$

(последний переход связан с тем, что $2^{k + 1} - 2^{k - 1} = 2\cdot2^k - 2^{k - 1} = 4\cdot2^{k - 1} - 2^{k - 1} = 3\cdot2^{k - 1}$).

Воспользуемся для вычисления предела правилом Лопиталя, выделим функции следующим образом: $f(k) = k$, $g(k) = \left(\frac{1}{4}\right)^{-3\cdot2^{k - 1}}$, рассчитаем их производные:

$$f^{'}(k) = 1$$

$$g^{'}(k) = \left(\frac{1}{4}\right)^{-3\cdot2^{k - 1}}\cdot\ln\frac{1}{4}\cdot(-3)\cdot2^{k - 1}\cdot\ln2 = -3\cdot\ln\frac{1}{4}\cdot\ln2\cdot\frac{2^{k - 1}}{\left(\frac{1}{4}\right)^{3\cdot2^{k - 1}}} = C\frac{2^{k - 1}}{(2^{-2})^{3\cdot2^{k - 1}}} =C\frac{2^{k - 1}}{(2^{2^{k - 1}\cdot 2})^{-3}} = C\frac{2^{k - 1}}{(2^{2^k})^{-3}} =  C\cdot2^{k - 1}\cdot2^{3\cdot2^k}$$

Тогда продолжим вычисление предела

$$(*)= \lim_{k\to\infty}\frac{f^{'}(k)}{g^{'}(k)} = \lim_{k\to\infty}\frac{1}{C\cdot2^{k - 1}\cdot2^{3\cdot2^k}} = 0\text{ - тоже сверхлинейно.}$$

8). Determine the convergence or divergence of a given sequence $r_k =\begin{cases} \frac{1}{k}, & \mbox{if } k\mbox{ is even} \\ \frac{1}{k^2}, & \mbox{if } k\mbox{ is odd} \end{cases}$.

##### Решение:

Аналогично предыдущему номеру разделим задачу на части; для чётных $k$:

$$\lim_{k\to\infty}\frac{r_{k + 1}}{r_k} = \lim_{k\to\infty}\frac{k}{(k + 1)^2} = \lim_{k\to\infty}\frac{k}{k^2 + 2k + 1} = \lim_{k\to\infty}\frac{1}{k + 2 + \frac{1}{k}} = 0\Rightarrow\text{сверхлинейная сходимость}$$

Для нечётных $k$:

$$\lim_{k\to\infty}\frac{r_{k + 1}}{r_k} = \lim_{k\to\infty}\frac{k^2}{k + 1} = \lim_{k\to\infty}\frac{1}{\frac{1}{k} + \frac{1}{k^2}} = \infty\Rightarrow\text{сходимости нет}$$

В итоге получается, что данная последовательность расходится.

9). Let $\left\{ r_k \right\}_{k=m}^\infty$ be a sequence of non-negative numbers and let $s > 0$ be some integer. Prove that sequence $\left\{ r_k \right\}_{k=m+s}^\infty$ is linearly convergent with constant $q$ if and only if a the sequence $\left\{ r_k \right\}_{k=m}^\infty$ converged linearly with constant $q$.
0

##### Решение:

$\Rightarrow:\;$ Пусть $\{r_k\}_{k=m+s}^{\infty}$ сходится линейно с $q$, распишем это по определению: $\exists C>0:\;\forall k\geq m+s\hookrightarrow r_k\leq Cq^k.$ Выберем константу $\widetilde{C}:=max\left\{C,\;\frac{r_m}{q^m}\;,\frac{r_{m+s-1}}{q^{m+s-1}}\right\}$, тогда $\forall k\geq m \hookrightarrow r_k\leq\widetilde{C}q^k$, что равносильно по определению тому, что $\{r_k\}_{k=m}^{\infty}$ тоже сходится линейно с $q$.

$\Leftarrow:\;$ $\{r_k\}_{k=m+s}^{\infty}$ является подпоследовательностью $\{r_k\}_{k=m}^{\infty}\Rightarrow$ если $\{r_k\}_{k=m}^{\infty}$ сходится линейно с $q$, то и любая её подпоследовательность тоже.

1. Consider the function $$f(x) = (x + \sin x) e^x, \;\;\; x \in [-20, 0]$$ 
    ![](https://raw.githubusercontent.com/MerkulovDaniil/optim/master/docs/exercises/Unimodal.svg)

    Consider the following modification of solution localization method, in which the interval $[a,b]$ is divided into $2$ parts in a fixed proportion of $t: x_t = a + t*(b-a)$ (maximum twice at iteration - as in the dichotomy method). Experiment with different values of $t \in [0,1]$ and plot the dependence of $N (t)$ - the number of iterations needed to achieve $\varepsilon$ - accuracy from the $t$ parameter. Consider $\varepsilon = 10^{-7}$. Note that with $t = 0.5$ this method is exactly the same as the dichotomy method.

```python
### YOUR CODE HERE

import warnings
warnings.filterwarnings('ignore')

from math import sin, exp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

f = lambda x: (x + sin(x)) * exp(x)

def LineSearch(f, t, a, b):
    """ Find minimum of the function (works good only in case of 1 minimum on the segment);
        Input: f - function to minimize, t - parameter in x = a + t * (b - a), a and b - bounds for x;
        Output: minimum x and number of iterations.
    """
    iter_counter = 1
    x = a + t * (b - a)
    while abs(b - a) >= 1e-7:
        iter_counter += 1
        left = a + t * (x - a)
        right = x + t * (b - x)
        if f(left) > f(right):
            a = x
            x = a + t * (b - a)
        else:
            b = x
            x = a + t * (b - a)
        #print(x)
    return x, iter_counter

a = -20
b = 0
T = np.linspace(0.01, 0.99, 50)
N = np.array([LineSearch(f, t, a, b)[1] for t in T])

plt.figure(figsize=(10, 5))
plt.plot(T, N)
plt.title('Dependence number of iterations N from the t parameter')
plt.xlabel('t')
plt.ylabel('N')
plt.show()
```

3. Show that the one-dimensional minimizer of a strongly convex quadratic function always satisfies the Goldstein conditions.

##### Решение:

Рассмотрим строго выпуклую квадратичную функцию $f(x) = \frac{1}{2}x^TAx - b^T,\;\;\nabla f(x) = Ax - b.$

Оптимальный шаг по методу наискорейшего спуска равняется $\alpha_k = -\frac{\nabla f_k^Td_k}{d_k^TAd_k} = -\frac{(x_k^TA - b^T)d_k}{d_k^TAd_k}.$

Условия Гольштейна: $f(x_k) + (1 − c)\alpha_k\nabla f_k^Td_k\leq f(x_k + \alpha_kd_k)\leq f(x_k) + c\alpha_k\nabla f_k^Td_k,\;\;\forall c\in(0,\frac{1}{2}).$

Распишем $f(x_k + \alpha_kd_k) = \frac{1}{2}x_kAx_k + \alpha_kx_k^TAd_k + \frac{1}{2}\alpha^2_kd_k^TAd_k - b^Tx_k - \alpha_kb^Td_k = f(x_k) + \alpha_kx_k^TAd_k + \frac{1}{2}\alpha_k^2d_k^TAd_k - \alpha b^Td_k$.

Также $(1 - c)\alpha_k\nabla f_k^Td_k = (1 - c)\alpha_k(x_k^TA - b^T)d_k = \alpha_k(x_k^TAd_k - cx^T_kAd_k - b^Td_k + cb^Td_k)$ и $x\alpha_k\nabla f_k^Td_k = \alpha_kc(x_k^TA - b^T)d_k,\;\; c\alpha_k\nabla f_k^Td_k = \alpha_kc(x_k^TA - b^T)d_k.$

Тогда второе условие Гольштейна можно записать в следующем виде: $\alpha_k(x_k^TAd_k - cx_k^TAd_k - b^Td_k + cb^Td_k)\leq\alpha_kx_k^TAd_k + \frac{1}{2}\alpha^2_kd_k^TAd_k - \alpha_kb^Td_k\leq\alpha_kc(x_k^TA - b^T)d_k.$

Рассмотрим первое условие и подставим в него значение оптимального шага: $-cx_k^TAd_k + cb^Td_k\leq\frac{1}{2}\alpha_kd_k^TAd_k\Rightarrow 2c(b^T - x_k^TA)d_k\leq (b^T - x_k^TA)d_k$ - это неравенство выполняется для $\forall c\in(0,\frac{1}{2}),$ если $(b^T - x_kA)d_k$ неотрицательно.

Для второго условия: $x_k^TAd_k + \frac{1}{2}\alpha_kd_k^TAd_k - b^Td_k\leq c(x_k^TA - b^T)d_k\Rightarrow \alpha_kd_k^TAd_k\leq 2(c - 1)(x_k^TA - b^T)d_k\Rightarrow (b^T - x_k^TA)d_k\leq 2(1 - c)(b^T - x_k^TA)d_k$ - верно при $2(1 - c)\geq 1$ (т.е. $c\leq\frac{1}{2}$) и неотрицательном $(b^T - x_k^TA)d_k.$

Implement Rastrigin function $f: \mathbb{R}^d \to \mathbb{R}$ for d = 10. [link](https://www.sfu.ca/~ssurjano/rastr.html)

$$
f(\mathbf{x})=10 d+\sum_{i=1}^{d}\left[x_{i}^{2}-10 \cos \left(2 \pi x_{i}\right)\right]

```python
import random
import numpy as np
from scipy.optimize import basinhopping, brute, differential_evolution, shgo, dual_annealing

def Rastrigin(X):
    return 10 * X.shape[0] + np.sum(X**2 - 10 * np.cos(2 * np.pi * X))

N2 = [[], [], [], [], []] # list for nfev of the following methods: 
                          # basinhopping, brute, differential_evolution, shgo, dual_annealing
N4 = [[], [], [], [], []] 
N8 = [[], [], [], [], []] 
N16 = [[], [], [], [], []] 
N32 = [[], [], [], [], []] 
f2 = [[], [], [], [], []]
f4 = [[], [], [], [], []]
f8 = [[], [], [], [], []]
f16 = [[], [], [], [], []]
f32 = [[], [], [], [], []]

def OneRun(niter):
    np.random.seed(random.randint(1, 200))
    
    x2 = np.random.randint(-5, 5, 2)
    x4 = np.random.randint(-5, 5, 4)
    x8 = np.random.randint(-5, 5, 8)
    x16 = np.random.randint(-5, 5, 16)
    x32 = np.random.randint(-5, 5, 32)
    
    # basinhopping
    basinhopping2 = basinhopping(Rastrigin, x2, niter=niter)
    basinhopping4 = basinhopping(Rastrigin, x4, niter=niter)
    basinhopping8 = basinhopping(Rastrigin, x8, niter=niter)
    basinhopping16 = basinhopping(Rastrigin, x16, niter=niter)
    basinhopping32 = basinhopping(Rastrigin, x32, niter=niter)
    
    print('kek1')

    # brute, only possible to compute for a function of two variables due to memory usage
    brute2 = brute(Rastrigin, [slice(-5, 5, 5/niter)] * 2, full_output=1)

    # differential_evolution
    differential_evolution2 = differential_evolution(Rastrigin, [(-5, 5)] * 2, maxiter=niter)
    differential_evolution4 = differential_evolution(Rastrigin, [(-5, 5)] * 4, maxiter=niter)
    differential_evolution8 = differential_evolution(Rastrigin, [(-5, 5)] * 8, maxiter=niter)
    differential_evolution16 = differential_evolution(Rastrigin, [(-5, 5)] * 16, maxiter=niter)
    differential_evolution32 = differential_evolution(Rastrigin, [(-5, 5)] * 32, maxiter=niter)
    
    print('kek2')

    # shgo
    shgo2 = shgo(Rastrigin, [(-5, 5)] * 2, n=niter)
    shgo4 = shgo(Rastrigin, [(-5, 5)] * 4, n=niter)
    shgo8 = shgo(Rastrigin, [(-5, 5)] * 8, n=niter)
    #shgo16 = shgo(Rastrigin, [(-5, 5)] * 16) за полчаса не сработало :(
    #shgo32 = shgo(Rastrigin, [(-5, 5)] * 32)
    
    print('kek3')

    # dual_annealing
    dual_annealing2 = dual_annealing(Rastrigin, [(-5, 5)] * 2, maxiter=niter)
    dual_annealing4 = dual_annealing(Rastrigin, [(-5, 5)] * 4, maxiter=niter)
    dual_annealing8 = dual_annealing(Rastrigin, [(-5, 5)] * 8, maxiter=niter)
    dual_annealing16 = dual_annealing(Rastrigin, [(-5, 5)] * 16, maxiter=niter)
    dual_annealing32 = dual_annealing(Rastrigin, [(-5, 5)] * 32, maxiter=niter)
    
    # add values to lists
    N2[0].append(basinhopping2.nfev)
    N4[0].append(basinhopping4.nfev)
    N8[0].append(basinhopping8.nfev)
    N16[0].append(basinhopping16.nfev)
    N32[0].append(basinhopping32.nfev)
    
    N2[1].append(brute2[3].size)
    
    N2[2].append(differential_evolution2.nfev)
    N4[2].append(differential_evolution4.nfev)
    N8[2].append(differential_evolution8.nfev)
    N16[2].append(differential_evolution16.nfev)
    N32[2].append(differential_evolution32.nfev)
    
    N2[3].append(shgo2.nfev)
    N4[3].append(shgo4.nfev)
    N8[3].append(shgo8.nfev)
    
    N2[4].append(dual_annealing2.nfev)
    N4[4].append(dual_annealing4.nfev)
    N8[4].append(dual_annealing8.nfev)
    N16[4].append(dual_annealing16.nfev)
    N32[4].append(dual_annealing32.nfev)

    f2[0].append(basinhopping2.fun)
    f4[0].append(basinhopping4.fun)
    f8[0].append(basinhopping8.fun)
    f16[0].append(basinhopping16.fun)
    f32[0].append(basinhopping32.fun)
    
    f2[1].append(brute2[1])
    
    f2[2].append(differential_evolution2.fun)
    f4[2].append(differential_evolution4.fun)
    f8[2].append(differential_evolution8.fun)
    f16[2].append(differential_evolution16.fun)
    f32[2].append(differential_evolution32.fun)
    
    f2[3].append(shgo2.fun)
    f4[3].append(shgo4.fun)
    f8[3].append(shgo8.fun)
    
    f2[4].append(dual_annealing2.fun)
    f4[4].append(dual_annealing4.fun)
    f8[4].append(dual_annealing8.fun)
    f16[4].append(dual_annealing16.fun)
    f32[4].append(dual_annealing32.fun)
    
    return 
```

## Hyperparameter search with optuna
Machine learning models often have hyperparameters. To choose optimal one between them one can use GridSearch or RandomSearch. But these algorithms computationally uneffective and don't use any sort of information about type of optimized function. To overcome this problem one can use [bayesian optimization](https://distill.pub/2020/bayesian-optimization/). Using this method we optimize our model by sequentially chosing points based on prior information about function. ![Image](https://www.resibots.eu/limbo/_images/bo_concept.png).

Write `conditional_SGD`, which returns the entire optimization trajectory $(w_k)$ after a fixed number of steps and draws the process on the graphs $P$ and $\nabla P$ that were above.

The conditional gradient descent method consists in making a gradient step and then checking if the obtained point belongs to the feasible set. If it belongs to the target set, the algorithm continues, otherwise a projection to the feasible set is made.

```python
def conditional_SGD(cond, dP, x, w0, mu, Nsteps, p=0.4):
    # cond - описание допустимого множества, задаётся в виде функции
    W = []
    W.append(w0)
    w = w0
    iters = 0
    for i in range(Nsteps):
        s = cond(dP(w[0], w[1], x))
        if dP(w[0], w[1], x) * (x - s) <= p:
            break
        w = (1 - mu) * w + mu * s # based on Frank-Wolfe algorithm
        W.append(w)
        iters += 1
    return W
```

For example, steepest descent algorithm will be plotted with the following code:

```python
from scipy.optimize import minimize_scalar

def steepest_descent(x_0, f, df, *f_params, df_eps = 1e-2, max_iter = 1000):
    iterations = []
    x = np.array(x_0)
    iterations.append(x)
    while np.linalg.norm(df(x, *f_params)) > df_eps and len(iterations) <= max_iter:
        res = minimize_scalar(lambda alpha: f(x - alpha * df(x, *f_params), *f_params))
        alpha_opt = res.x
        x = x - alpha_opt * df(x, *f_params)
        iterations.append(x)
    print(f'Finished with {len(iterations)} iterations')
    return iterations

x_0 = [4,1]
k = 30
iterations = steepest_descent(x_0, f, df, k, df_eps = 1e-9)
title = f'$f(x_1, x_2) = x_1^2 + k x_2^2, k = {k}$'

plot_3d_function(x1, x2, f, title, k, minima=[0,0], iterations = iterations)
```

На всех графиках наблюдается четкое различие количества итераций для $k<1$ и $k>1$: в первом случае количество итераций значительно больше, чем во втором (и иногда даже превышает лимит). Для того, чтобы понять, почему так происходит, рассмотрим шаг градиентного спуска:

$\begin{pmatrix}
  x^{(k + 1)}_1\\
  x^{(k + 1)}_2
\end{pmatrix}=
\begin{pmatrix}
  x^{(k)}_1\\
  x^{(k)}_2
\end{pmatrix} - \alpha
\begin{pmatrix}
  2x^{(k)}_1\\
  2kx^{(k)}_2
\end{pmatrix}$

Видим, что минимизация по координате $x_2$ определяется не только значением шага $\alpha$, но и коэффициентом $k$. Поэтому, при маленьких $k$ "продвижение" к минимуму происходит крайне медленно и может даже не доходить до него, превышая лимит - как это происходит в левых частях представленных графиков. Наоборот, при увеличении $k$ продвижение к минимуму идёт быстрее за счёт увеличения значения $k\cdot\alpha$; соответственно, алгоритм работает за меньшее число итераций.

Заметим, что при увеличении $\alpha$ переход через $k=1$ становится более плавным; это, опять же, связано с увеличением значения произведения $\alpha k$ - продвижение к минимуму происходит быстрее при одних и тех же $k$ за счёт $\alpha$.

Интересен также выброс, который находится в правых частях всех построенных графиков. Он соответствует случаю, когда $\alpha k = 1$ - в таком случае для координаты $x_2$ на всех итерациях градиентного спуска мы получаем значения либо $x^{(0)}_2$, либо $-x^{(0)}_2$ (где $x^{(0)}_2$ - начальное приближение) - т.е. шагаем из стороны в сторону по двум значениям для $x_2$, продвигаясь только по $x_1$; из-за этого процесс оптимизации усложняется и количество итераций резко возрастает.

## Projected gradient descent

Find projection on the $S$ set $\pi_S (y) = \pi$ if: 

$$
S = \{x \in \mathbb{R}^n \mid Ax = b, A \in \mathbb{R}^{m \times n}, b \in \mathbb{R}^{m}\}, y \notin S
$$

Hint: Check [fmin.xyz](https://fmin.xyz) 

<font color="lime"> ==YOUR ANSWER== </font>

Нахождение проекции эквивалентно следующей задаче:
$$
\left\{ \begin{array}{l} 
\displaystyle \frac{1}{2}\big\lVert x - y \big\rVert_2^2 \longrightarrow \min_{x \in \mathbb{R}^n}, \\\\[-1em]
Ax = b.
\end{array} \right.
$$

(первое условие из того, что расстояние от точки до проекции должно быть минимально, а второе из того, что $\pi\in S$).

Решим эту задачу:

составим функцию Лагранжа $L = \frac{1}{2}||x - y||^2 + \alpha ^T(Ax - b)$,

найдём оптимальное значение $\nabla L = y - x^* + A^T\alpha = 0\Rightarrow x^* = \pi = y + A^T\alpha\Rightarrow A\pi = Ay + AA^T\alpha$ или, используя второе уравнение системы, $Ay = b - AA^T\alpha$.

$\;\;$

For following problem:

$$
\left\{ \begin{array}{l} 
\displaystyle \big\lVert Ax - b \big\rVert_2^2 \longrightarrow \min_{x \in \mathbb{R}^n}, \\\\[-1em]
Cx = d.
\end{array} \right. \ \ \begin{array}{l}
A \in \mathbb{R}^{m \times n}, \ m \geq n, \ \text{rg}\, A = n \\
C \in \mathbb{R}^{k \times n}, \ k \leq n, \ \text{rg}\, C = k
\end{array} \ b \in \mathbb{R}^m, \ d \in \mathbb{R}^k
$$

* Write KKT conditions;
* Find $x^\ast$ - solution;

<font color="lime"> ==YOUR ANSWER== </font>

1). KKT:

$Cx - d = 0$\
$\lambda \geq 0$\
$2A^T(Ax - b) + C^T\lambda = 0$

$\;\;$

2). Найдём $x^*$:

Функция Лагранжа $L(x,\lambda) =  (Ax − b)^T(Ax − b) + \lambda^T(Cx − d) = x^TA^TAx + (C^T\lambda − 2A^Tb)^T x − \lambda ^Td + b^Tb$

Ещё раз ККТ:

$\nabla _x L = 2A^TAx^* + C^T\lambda^* − 2A^Tb = 0$

$\nabla _{\lambda} L = Cx^* - d = 0$.

Выразим из первого условия $x^* = (A^TA)^{-1}(A^Tb - \frac{1}{2}C^T\lambda^*)$ -- обратная матрица для $A^TA$ существует, т.к. $rank(A) = n$ - равен кол-ву столбцов в ней, т.е. $A^TA$ точно невырожденная).

Подставим найденный $x^*$ во второе условие: $C(A^TA)^{−1}A^Tb − \frac{1}{2}C(A^TA)^{−1}C^T\lambda ^* = d$

и выразим $\lambda ^* = −2(C(A^TA)^{−1}C^T)^{−1}(d − C(A^TA)^{−1}Ab$ -- обратная матрица для $C(A^TA)^{−1}C^T$ существует по аналогичным причинам.

Тогда подставим $\lambda ^*$ в предыдущее выражение для $x^*$:

$x^* = (A^TA)^{−1}(A^Tb + C^T(C(A^TA)^{−1}C^T)^{−1}(d − C(A^TA)^{−1}A^Tb))$.

$\;\;$



Implement projected gradient descent for following task.

$$
\left\{ \begin{array}{l} 
\displaystyle \big\lVert Ax - b \big\rVert_2^2 \longrightarrow \min_{x \in \mathbb{R}^n}, \\\\[-1em]
Cx = d.
\end{array} \right. \ \ \begin{array}{l}
A \in \mathbb{R}^{m \times n}, \ m \geq n, \ \text{rg}\, A = n \\
C \in \mathbb{R}^{k \times n}, \ k \leq n, \ \text{rg}\, C = k
\end{array} \ b \in \mathbb{R}^m, \ d \in \mathbb{R}^k
$$

Compare with analytical solution and `scipy.optimize`

```python
import numpy as np
```

```python
np.random.seed(1)

A = np.random.randn(100, 20)
b = np.random.randn(100)

C = np.random.randn(10, 20)
d = np.random.randn(10)
```

```python
def f(x):
    ### YOUR CODE
    return np.square(np.linalg.norm(np.dot(A, x) - b))

def df(x):
    return 2 * np.dot(A.transpose(), (np.dot(A, x) - b))
```

```python
def proj(C, d, x):
    """ Project x on the subspace Cx = d using Singular Decomposition;
        Input: C, d (from definition of the subspace) and vector x to project
        Output: projection x
    """
    k, n = C.shape
    u, s, vh = np.linalg.svd(C) # s - vector of singular values
    eps = np.finfo(float).eps # machine limits for floating point types
    bad = (s < s[0] * eps)
    i = bad.searchsorted(1)
    if i < k:        # for bad-conditioned C, find the first i such that s[i]/s[0] < eps
        rcond = s[i]
    else:            # for a well-conditioned C
        rcond = -1
    x0 = np.linalg.lstsq(C, d, rcond=rcond)[0] # rcond is a cut-off ratio for small singular values of C
    null_space = vh[i:]
    return x0 + (null_space * np.dot(null_space, (x - x0))[:,np.newaxis]).sum(axis=0)
```

```python
x = np.zeros(20)
### Projected gradient descent
def projected_gradient_descent(x0, df, mu, Nsteps):
    x = x0
    for i in range(Nsteps):
        x = proj(C, d, x - mu * df(x))
    return x

ans_proj = projected_gradient_descent(x, df, 0.001, 732)
```

Comparison with `scipy.optimize`

```python
### YOUR CODE
from scipy.optimize import minimize
from scipy.optimize import LinearConstraint

ans_scipy = minimize(f, x, jac=df, constraints = LinearConstraint(C , d, d)).x
ans_scipy
```

Comparison with analytical solution

$x^* = (A^TA)^{−1}(A^Tb + C^T(C(A^TA)^{−1}C^T)^{−1}(d − C(A^TA)^{−1}A^Tb))$.


```python
### YOUR CODE
from numpy.linalg import inv

R1 = inv(A.transpose() @ A)
R2 = C.transpose() @ inv(C @ inv(A.transpose() @ A) @ C.transpose())
R3 = d - C @ inv(A.transpose() @ A) @ A.transpose() @ b

ans_analit = R1 @ (A.transpose() @ b + R2 @ R3)
ans_analit
```

```python
display(np.around(ans_analit, 14) == np.around(ans_proj, 14))
display(np.around(ans_analit, 3) == np.around(ans_scipy, 3))
display(np.around(ans_scipy, 3) == np.around(ans_proj, 3))
```
